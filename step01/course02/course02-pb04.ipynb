```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'  # 또는 다른 한글 폰트
plt.rcParams['axes.unicode_minus'] = False    # 마이너스 기호 깨짐 방지
```

```python
# 데이터 로드
genres_df = pd.read_csv('data/genres.csv')
movies_df = pd.read_csv('data/movies.txt', sep='\t')
rates_df = pd.read_csv('data/rates.csv')
```

```python
import seaborn as sns
# 스타일 설정
sns.set(font='NanumGothic', # 한글 폰트
        rc={'figure.figsize':(12,8)}, # 그래프 크기
        style='whitegrid'    # 그래프 스타일
)
```

```python
# 장르별 영화 수 계산
genre_counts = genres_df['genre'].value_counts()

# 장르별 분포 시각화 (Bar chart)
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.title('Distribution of Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 장르와 평점 데이터 병합
merged_df = pd.merge(genres_df, rates_df, on='movie')

# 장르별 평균 평점 계산
genre_avg_rates = merged_df.groupby('genre')['rate'].mean().sort_values(ascending=False)

# 장르별 평균 평점 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_avg_rates.index, y=genre_avg_rates.values)
plt.title('Average Ratings by Movie Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
